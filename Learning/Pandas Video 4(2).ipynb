```python
import pandas as pd
import numpy as np


data = {
    'Product': ['A', 'B', 'C', 'D', '7', 'F'],
    'Revenue': [5000, -1500, 7000, -3000, 12000, 8000],
    'Category': ['Electronics', 'Clothing', 'Grocery', 'Clothing', 'Electronics', 'Grocery']
}

df = pd.DataFrame(data)
df.replace({'Revenue': { x: np.nan for x in df['Revenue'] if x<0}}, inplace = True)
def is_numeric(s):
    try:
       float(s)
       return True
    except (ValueError, TypeError):
       return False
df['Product'] = df['Product'].apply(lambda x: np.nan if pd.notna(x) and is_numeric(x) else x)
df

#This is an anonymous function (lambda function) applied to each value x in the column.
#pd.notna(x): Checks if x is not NaN (ensures we only process actual values).
#is_numeric(x): Presumably a function that checks if x is a numeric value.
#If x is numeric and not NaN, it is replaced with np.nan.
```
